```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator, FuncFormatter, NullLocator

plt.style.use('pablet')
```

```python
data1= pd.read_csv('Alavi.dat', sep='\s+')
M1= data1['M1']
phi1= data1['phi1']*1e-2
error_l1= data1['err_l1']*1e-2
error_h1= data1['err_h1']*1e-2
error1= [error_l1, error_h1]
M2= data1['M2']
phi2= data1['phi2']*1e-2
error_l2= data1['err_l2']*1e-2
error_h2= data1['err_h2']*1e-2
error2= [error_l2, error_h2]
M3= data1['M3']
phi3= data1['phi3']*1e-2
error_l3= data1['err_l3']*1e-2 
error_h3= data1['err_h3']*1e-2
error3= [error_l3, error_h3]

data2= pd.read_csv('Bouwens.dat', sep='\s+')
M4= data2['M4']
phi4= data2['phi4']
error4= data2['err4']
M5= data2['M5']
phi5= data2['phi5']
error5= data2['err5']
M6= data2['M6']
phi6= data2['phi6']
error6= data2['err6']
M7= data2['M7']
phi7= data2['phi7']
error7= data2['err7']
M8= data2['M8']
phi8= data2['phi8']
error8= data2['err8']
M10= data2['M10']
phi10= data2['phi10']
error10= data2['err10']
```

```python
def Schechter(M, A, M_c, alpha):
    return A*(np.log(10)/2.5)*10**(-0.4*(M-M_c)*(alpha+1))*np.exp(-10**(-0.4*(M-M_c)))
def M(mag):
    return np.linspace(mag.max()+0.5, mag.min()-0.5, 100)

A= np.asarray([1.97,0.74,0.50,0.29,0.21,0.008])/1e3
M_c= [-20.88,-21.17,-20.94,-20.87,-20.63,-20.92]
alpha= [-1.64,-1.76,-1.87,-2.06,-2.02,-2.27]

A2= np.asarray([2.21,1.46,0.48])/1e3
M_c2= [-19.85,-20.39,-20.70]
alpha2= [-1.50,-1.80,-2.01]
```

```python
fig= plt.figure()

plt.errorbar(M1, phi1, yerr=error1, fmt='o', capsize=5, ms=5, c='#6e7bf0')
plt.plot(M(M1), Schechter(M(M1), A2[0], M_c2[0], alpha2[0]), label='z ~ 1', linewidth=2, c='#6e7bf0', alpha=0.5)
plt.errorbar(M2, phi2, yerr=error2, fmt='o', capsize=5, ms=5, c='#a31443')
plt.plot(M(M2), Schechter(M(M2), A2[1], M_c2[1], alpha2[1]), label='z ~ 2', linewidth=2, c='#a31443', alpha=0.5)
plt.errorbar(M3, phi3, yerr=error3, fmt='o', capsize=5, ms=5, c='#c3c906')
plt.plot(M(M3), Schechter(M(M3), A2[2], M_c2[2], alpha2[2]), label='z ~ 3', linewidth=2, c='#c3c906', alpha=0.5)
plt.errorbar(M4, phi4, yerr=error4, fmt='o', capsize=5, ms=5, c='#0008ff')
plt.plot(M(M4), Schechter(M(M4), A[0], M_c[0], alpha[0]), label='z ~ 4', linewidth=2, c='#0008ff', alpha=0.5)
plt.errorbar(M5, phi5, yerr=error5, fmt='o', capsize=5, ms=5, c='#11ff00')
plt.plot(M(M5), Schechter(M(M5), A[1], M_c[1], alpha[1]), label='z ~ 5', linewidth=2, c='#11ff00', alpha=0.5)
plt.errorbar(M6, phi6, yerr=error6, fmt='o', capsize=5, ms=5, c='#64bdb8')
plt.plot(M(M6), Schechter(M(M6), A[2], M_c[2], alpha[2]), label='z ~ 6', linewidth=2, c='#64bdb8', alpha=0.5)
plt.errorbar(M7, phi7, yerr=error7, fmt='o', capsize=5, ms=5, c='#000000')
plt.plot(M(M7), Schechter(M(M7), A[3], M_c[3], alpha[3]), label='z ~ 7', linewidth=2, c='#000000', alpha=0.5)
plt.errorbar(M8, phi8, yerr=error8, fmt='o', capsize=5, ms=5, c='#ff1500')
plt.plot(M(M8), Schechter(M(M8), A[4], M_c[4], alpha[4]), label='z ~ 8', linewidth=2, c='#ff1500', alpha=0.5)
plt.errorbar(M10, phi10, yerr=error10, fmt='o', capsize=5, ms=5, c='#bf00ff')
plt.plot(M(M10), Schechter(M(M10), A[5], M_c[5], alpha[5]), label='z ~ 10', linewidth=2, c='#bf00ff', alpha=0.5)

plt.yscale('log')
plt.ylabel(r'$\log_{10}(\Phi/$ Mpc$^{-3}$ mag$^{-1})$')
plt.xlabel(r'M$_{AB}$')

def exponent_formatter(val, pos):
    return f"{int(np.log10(val))}"
plt.gca().yaxis.set_major_locator(LogLocator(base=10.0))
plt.gca().yaxis.set_major_formatter(FuncFormatter(exponent_formatter))
plt.gca().yaxis.set_minor_locator(NullLocator())

plt.xlim(-23.4, -12)
plt.ylim(5e-7, 2)
plt.legend();
```
