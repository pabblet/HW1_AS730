```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('pablet')
```

```python
data= pd.read_csv('SDSS_SQL.csv', sep=',', skiprows=1)
data.info()
z= np.asarray(data['z'])
M= np.asarray(data['stellarM'])
SFR= np.asarray(data['SFR'])
R= np.asarray(data['absMagR'])
G= np.asarray(data['absMagG'])
```

```python
plt.figure()
scatter= plt.scatter(R, G-R, c=M, cmap='viridis', s=5)
sns.kdeplot(x=data['absMagR'], y=data['absMagG']-data['absMagR'], levels=7, color='#dedede', linewidths=2)

plt.title('Color Magnitude Diagram for galaxies at z$\lesssim$0.1')
plt.colorbar(label='log$_{10}$(M$_*$)')
plt.ylabel('$M_g-M_r$');
plt.xlabel('$M_r$');
plt.ylim(-1,2);
plt.xlim(-15,-24.5);
```

```python
plt.figure()
scatter= plt.scatter(data['absMagR'], data['absMagG']-data['absMagR'], c=np.log10(10**data['SFR']/10**data['stellarM']), cmap='viridis', s=5)
sns.kdeplot(x=data['absMagR'], y=data['absMagG']-data['absMagR'], levels=7, color='#dedede', linewidths=2)

plt.title('Color Magnitude Diagram for galaxies at z$\lesssim$0.1')
plt.colorbar(label='log$_{10}$(sSFR)')
plt.ylabel('$M_g-M_r$');
plt.xlabel('$M_r$');
plt.ylim(-1,2);
plt.xlim(-15,-24.5);
```
