```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('pablet')
```

```python
gal= 'Arp22'
data= pd.read_csv(gal+'.csv', sep='\s+', names=['wl', 'flux', 'a', 'b'])
wl= data['wl']
flux= data['flux']
print(data)
```

```python
plt.plot(wl, flux, c='#444444')
plt.title(gal)
plt.ylabel(r'F$_\lambda$ [ergs/s/cm$^2$/$\AA$]')
plt.xlabel(r'Wavelength [$\AA$]')
plt.yscale('log');
#plt.xscale('log');
```
